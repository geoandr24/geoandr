```python
import pandas as pd
df=pd.read_csv('India_GDP_1960-2022.csv')
```

```python
print(df)
df.info()
```

```python
print(df)

df['gdp_billion_dollars']=pd.to_numeric(df['gdp_billion_dollars'], errors='coerce')
df['per_capita_in_rupees']=pd.to_numeric(df['per_capita_in_rupees'], errors='coerce')
df['growth']=pd.to_numeric(df['growth'], errors='coerce')
new_row={'year':'2021','gdp_billion_dollars':3173,'per_capita_in_rupees':182160,'growth':8.95}
df=df.append(new_row,ignore_index=True)
print(df)
```

```python
print(df.info())
df['per_capita_in_dollars']=[row*0.012 for row in df['per_capita_in_rupees']]
print(df)

df=df.sort_values(by='year',ascending=True)
```

```python
from matplotlib import pyplot as plt
ax=plt.subplot()
plt.plot(df.year,df.gdp_billion_dollars)
plt.plot(df.year,df.per_capita_in_dollars)
ax.set_xticks(['1960','1970','1980','1990','2000','2010','2021'])
plt.xlabel('Years')
plt.legend(['GDP','GDP per capita'])
```

```python
#Create 2 y-axis
from matplotlib import pyplot as plt
col1='steelblue'
col2='orange'
plt.figure(figsize=(30,30))

fig,ax=plt.subplots()
ax.plot(df.year,df.gdp_billion_dollars,color=col1)

ax.set_xlabel('Year',fontsize=12)
ax.set_xticks(['1960','1970','1980','1990','2000','2010','2021'])

ax.set_ylabel('GDP in billions $',fontsize=12,color=col1)

ax2=ax.twinx()

ax2.plot(df.year,df.per_capita_in_dollars,color=col2)

ax2.set_ylabel('GDP per capita $',color=col2,fontsize=12)
```

```python
ax=plt.subplot()
plt.plot(df.year,df.growth)
ax.set_xticks(['1960','1970','1980','1990','2000','2010','2021'])
plt.xlabel('Years')
plt.ylabel('Growth')
plt.show()
```

```python
df2=pd.read_csv('avg.csv')
print(df2)
df2.columns=['avg_growth_rate']
df['avg_growth_rate']=df2
```

```python
print(df)
```

```python
plt.figure(figsize=(10,5))
ax=plt.subplot()
plt.plot(df.year,df.growth)
plt.plot(df.year,df.avg_growth_rate)

ax.set_xticks(['1960','1970','1980','1990','2000','2010','2021'])
plt.legend(['Growth Rate','Average Growth Rate'],loc=3)
plt.xlabel('Years',fontsize=12)
plt.ylabel('Growth',fontsize=12)
plt.show()

```
